Run multi scripts parallel

```python
from concurrent.futures import ThreadPoolExecutor
import subprocess
import os
scripts = [
    r"D:\Maged\Courses\Data Engineering\Lec2\Urban Services Intelligence\gas_stations_scraping.py",
    r"D:\Maged\Courses\Data Engineering\Lec2\Urban Services Intelligence\hospitals_scraping.py",
    r"D:\Maged\Courses\Data Engineering\Lec2\Urban Services Intelligence\schools_scraping.py",
    r"D:\Maged\Courses\Data Engineering\Lec2\Urban Services Intelligence\university_scraping.py",
    r"D:\Maged\Courses\Data Engineering\Lec2\Urban Services Intelligence\clubs_scraping.py",
]

def run_script(script):
    print(f"🟢 تشغيل السكريبت: {os.path.basename(script)}")
    
    env = os.environ.copy()
    env["PYTHONIOENCODING"] = "utf-8"
    
    process = subprocess.Popen(
        ["python", script],
        stdout=subprocess.PIPE,
        stderr=subprocess.STDOUT,
        bufsize=1,
        text=True,
        encoding="utf-8",   # ← هنا مهم
        env=env
    )
    
    for line in process.stdout:
        print(f"[{os.path.basename(script)}] {line.strip()}")
    
    process.wait()
    print(f"✅ انتهى السكريبت: {os.path.basename(script)}")

with ThreadPoolExecutor(max_workers=len(scripts)) as executor:
    executor.map(run_script, scripts)
```

```python
from matplotlib import category
import pandas as pd
from sqlalchemy import create_engine
from sqlalchemy.types import NVARCHAR, Float

category = ["clubs", "gas_stations", "hospitals", "schools", "universities"]

for c in category:
    # Connect to staging_db
    server = "localhost"
    database = "staging_c_db"

    connection_string = (
        f"mssql+pyodbc://@{server}/{database}"
        "?driver=ODBC+Driver+18+for+SQL+Server"
        "&TrustServerCertificate=yes"
    )
    engine = create_engine(connection_string)

    # Load Excel file
    df = pd.read_csv(
        fr"D:\Maged\Courses\Data Engineering\Lec2\Urban Services Intelligence\Data Sources\{c}.csv",encoding="utf-8")
    # Insert into staging table
    df.to_sql(
        f"stg_{c}",
        engine,
        if_exists="replace",
        index=False,
        dtype={
            "City": NVARCHAR(100),
            "Category": NVARCHAR(100),
            "Sub Category": NVARCHAR(200),
            "English Name": NVARCHAR(200),
            "Arabic Name": NVARCHAR(200),
            "Reviews": NVARCHAR(200),
            "Address": NVARCHAR(500),
            "Rating": Float
        }
    )
    print("CSV data loaded into staging_db!")
```
